Pandas 数据读取：
1. 数据类型：csv、tsv、txt等用逗号分隔、tab键分隔、空格分隔等格式存储的数据文件。读取方式：pd.read_csv()等。
2. 数据类型：excel等，读取方式：pd.read_excel()。
3. 数据类型：mysql等，读取方式：pd.read_sql()。链接数据库需要 pymysql 模块。

```python
import pandas as pd

data = pd.read_csv('student.csv',
                   names=['name', 'age', 'score'],  # 自定义列名
                   header=None,
                   )  
print(data)
print(data.head()) # 查看前5行数据
print(data.tail()) # 查看后5行数据
print(data.shape) # 查看数据维度
print(data.columns) # 查看数据列名
```

Pandas的数据类型：
1. Series：一维数组，类似列表，可以包含任何数据类型，如整数、浮点数、字符串等。
2. DataFrame：二维数组，类似表格，可以包含多个Series，每个Series代表一个列。
3. Panel：三维数组，类似立方体，可以包含多个DataFrame，每个DataFrame代表一个层。

```python
import pandas as pd
import numpy as np

s1 = pd.Series([1, 2, "Hello", 4])
print(s1)
print(s1.index)
s2 = pd.Series([1, 2, 3, 4], index=['a', 'b', 'c', 'd']) # 自定义索引
print(s2)
```

```python
data = pd.DataFrame({'name': ['Alice', 'Bob', 'Charlie'],
                     'age': [25, 30, 28],
                     'city': ['New York', 'Los Angeles', 'Chicago']
                     }) # 创建DataFrame
print(data) 
print(data.dtypes) # 查看数据类型
print(data.index) # 查看索引
print(data.columns) # 查看列名
print(data['name']) # 查看name列
print(data.loc[1]) # 查看第二行数据
print(data.iloc[1:3]) # 查看第二到第三行数据
```

Pandas对数据查询：
1. loc：根据标签查询数据，如data.loc[1]，表示查询第二行数据。
2. iloc：根据位置查询数据，如data.iloc[1:3]，表示查询第二到第三行数据。
3. where：根据条件查询数据，如data.where(data['age'] > 25)，表示查询年龄大于25的数据。
4. query：根据表达式查询数据，如data.query('age > 25')，表示查询年龄大于25的数据。
