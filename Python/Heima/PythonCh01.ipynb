变量：变量是一个内存地址，用来存储数据。
变量名 = 数据

数据类型转换：
   1. int(变量名)

标识符的命名规则：
1. 标识符只能包含字母、数字、下划线。
2. 标识符不能以数字开头。
3. 标识符区分大小写。
4. 标识符不能使用关键字。
5. 标识符不能使用保留字。
6. 标识符不能使用特殊字符。
7. 标识符不能使用空格。
8. 标识符不能使用中文。
9. 标识符不能使用特殊符号。

变量的命名规范： 
1. 见名知意
2. 下划线命名法
3. 字母全小写

运算符：
1. 算术运算符： + - * / % ** //
2. 比较运算符： > < >= <= == !=
3. 逻辑运算符： and or not
4. 赋值运算符： = += -= *= /= %= **= //=
5. 位运算符： & | ^ ~ << >>
6. 身份运算符： is, is not
7. 成员运算符： in, not in

```python
print("1 + 2 =", 1 + 2)
print("1 - 2 =", 1 - 2)
print("1 * 2 =", 1 * 2)
print("1 / 2 =", 1 / 2)
print("1 % 2 =", 1 % 2)
print("1 ** 2 =", 1 ** 2)
print("1 // 2 =", 1 // 2)
```

字符串的三种定义方式：
1. 单引号： 'abc'
2. 双引号： "abc"
3. 三引号： '''abc''' (支持多行字符串)

单引号表示可以内嵌双引号，双引号表示可以内嵌单引号。
或者使用 \ 转义符。

字符串的拼接：
1. 使用 + 注意拼接符只能用于字符串。
2. 占位符： %s %d %f %% 例如： "Hello %d + %d = %d, %s" % (1, 2, 3, "World")  输出： Hello 1 + 2 = 3, World
   1. %d 表示数字，可以控制宽度。
   2. %s 表示字符串
   3. %% 表示百分号
   4. %f 表示浮点数，默认显示小数点后6位。也可以显示小数点后指定位数。
   5. %c 表示字符
3. 格式化字符串： f"Hello {a} Python {b}"


```python
a = str("Hello")
b = str("World")
print(a + " Python " + b)
print("a = %d, b = %d" % (1, 2))
c = 1.3999999
print("c = %.2f" % c)
print(f"Hello {a} Python {b}")
```

数据输入： input() 函数    
输入的内容默认为字符串类型。可以使用强制类型转换。
1. int(input())
input() 中可以有参数，提示用户输入。
2. input("请输入：")
3. 此外 print() 函数默认换行，可以使用 print(,end = "") 来控制不换行。 

```python
name = str(input("请输入你的姓名："))
print(f"你的姓名是：{name}", end = "")
```
