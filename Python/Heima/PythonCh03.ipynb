变量的作用域:
1. 局部变量：定义在函数体内部的变量，只在函数内部起作用
2. 全局变量：将变量定义在函数的外边，可以在函数内部声明 global 关键字

```python
from dataclasses import field

num = 100

def a():
    print(num)
    
def b():
    num = 500
    print(num)
    
def c():
    global num
    print(num)
    num = 600
    print(num)
    
a()
b()
c()
print(num)
```

List:
1. list 是一个有序的集合，可以随时添加和删除其中的元素
2. List用括号括起来，元素之间用逗号隔开，L = [1,2,3]
3. list 的元素可以是不同类型，list 可以嵌套，list 可以作为参数传递给函数
4. list 可以通过索引访问，索引从0开始，负数索引从-1开始，-1是最后一个元素，-2是倒数第二个元素
5. list 可以通过切片访问，切片是左闭右开，L[1:3]表示从索引1到索引2，左闭右开
6. list 可以通过索引修改元素，L[1] = 2

```python
mylist = [1,2,3]
print(mylist)

mylist = [1,"Hello",True]
print(mylist)

mylist = [1,2,3,4,5,6,7,8,9]
print(mylist[0])
print(mylist[-1])
print(mylist[1:3])
mylist[1] = 2
print(mylist[:-1])

mylist = [[1,2,3], [4,5,6]]
print(mylist[0])
print(mylist[0][1])
```

List的常用操作:
1. append()：在列表末尾添加一个元素
2. extend()：将一个列表的元素添加到另一个列表中
3. insert()：在列表中插入一个元素
4. remove()：从列表中移除第一个指定元素
5. pop()：移除列表中的一个元素，默认为最后一个元素，并返回该元素
6. clear()：清空列表中的所有元素
7. index()：返回元素在列表中的索引
8. count()：返回元素在列表中出现的次数
9. sort()：对列表进行排序
10. reverse()：对列表进行逆序排列
11. del ：删除列表中的元素

```python
mylist = [1,2,3,4,5,6,7,8,9]
print(mylist.index(1)) # 输出元素 1的索引
print(mylist.count(1)) # 输出元素 1出现的次数
mylist.append(10) # 在列表末尾添加一个元素
print(mylist)
mylist.extend([11,12]) # 将一个列表的元素添加到另一个列表中
print(mylist)
mylist.insert(1,2) # 在列表中插入一个元素，在索引为1的位置插入元素2
print(mylist)
mylist.remove(2) # 从列表中移除第一个指定元素
print(mylist)
end = mylist.pop() # 移除列表中的一个元素，默认为最后一个元素，并返回该元素，也可以指定下标
print(mylist)
print(end)
del mylist[0] # 删除列表中的第一个元素
print(mylist)
mylist.remove(2) # 删除列表中的元素,但不返回
print(mylist)

import random

random.shuffle(mylist) # 将列表中的元素随机排序
print(mylist)
mylist.sort() # 对列表进行排序
print(mylist)
print(len(mylist))
mylist.reverse()
print(mylist)
mylist.clear()
print(mylist)
```

元组：
1. 元组和列表类似，但是元组不可变，一旦创建就不能修改
2. 元组用小括号括起来，元素之间用逗号隔开，t = (1,2,3)
3. 元组可以嵌套，元组可以作为参数传递给函数
4. 元组可以通过索引访问，索引从0开始，负数索引从-1开始，-1是最后一个元素，-2是倒数第二个元素
5. 元组可以通过切片访问，切片是左闭右开，t[1:3]表示从索引1到索引2，左闭右开
6. 元组可以通过索引修改元素，但是元组是不可变的，所以不能修改元组中的元素

```python
tpl = (1,2,3)
print(tpl)

tpl = (1,"Hello",True)
print(tpl)

# tpl[1] = 2 # 元组是不可变的，所以不能修改元组中的元素
tpl = (1,) # 元组只有一个元素，需要加逗号
print(tpl)
```

元组操作：
1. count()：返回元素在元组中出现的次数
2. index()：返回元素在元组中的索引
3. len()：返回元组长度

```python
tpl = (1,2,3,4,5,6,7,8,9)
print(tpl.count(1))
print(tpl.index(1))
print(len(tpl))

for i in tpl:
    print(i,end = "\t")
```

字符串：str()
1. 字符串是不可变的，一旦创建就不能修改
2. 字符串用单引号或双引号括起来，字符串可以嵌套，字符串可以作为参数传递给函数
3. 字符串可以通过索引访问，索引从0开始，负数索引从-1开始，-1是最后一个元素，-2是倒数第二个元素
4. 字符串可以通过切片访问，切片是左闭右开，s[1:3]表示从索引1到索引2，左闭右开
5. 字符串可以通过索引修改元素，但是字符串是不可变的，所以不能修改字符串中的元素
字符串的常用操作：
6. len()：返回字符串的长度
7. index(): 返回字符串中第一个指定字符的索引
8. count(): 返回字符串中指定字符出现的次数
9. find(): 返回字符串中第一个指定字符的索引，如果没有找到，返回-1
10. replace(): 返回字符串中第一个指定字符替换为指定字符
11. split(): 返回字符串以指定字符为分隔符，分割为多个字符串
12. strip(): 返回字符串去除两端的空白字符
13. lstrip(): 返回字符串去除左边的空白字符
14. rstrip(): 返回字符串去除右边的空白字符
15. lower(): 返回字符串转换为小写字母
16. upper(): 返回字符串转换为大写字母
17. title(): 返回字符串每个单词的首字母大写
18. swapcase(): 返回字符串每个字母的大小写反转
19. center(): 返回字符串居中，并使用指定字符填充
20. ljust(): 返回字符串左对齐，并使用指定字符填充
21. rjust(): 返回字符串右对齐，并使用

```python
string = str("           Hello World!")
print(string)
print(string.index("o"))
print(string.count("o"))
print(string.find("o"))
print(string.replace("o","O"))
print(string.split(" "))
print(string.strip())
print(string.lstrip())
print(string.rstrip())
print(string.lower())
print(string.upper())
print(string.title())
print(string.swapcase()) # 大小写反转
print(string.center(20,"%")) # 居中，并使用*填充
print(string.ljust(20,"%"))
print(len(string))
print(string.replace(" ","|"))
```

切片：[start:end:step]
1. 切片是左闭右开的，s[1:3]表示从索引1到索引2，左闭右开
2. 切片可以省略开始和结束索引，默认从0开始，到字符串末尾结束
3. 切片可以同时指定步长，s[1:3:2]表示从索引1到索引3，步长为2，每隔一个取一个
4. 切片可以同时指定开始和结束索引，并且步长为负数，s[::-1]表示从字符串末尾开始，到开头结束，步长为-1，逆序输出字符串
5. 切片可以同时指定开始和结束索引，并且步长为正数，s[::1]表示从开头开始，到结尾结束，步长为1，正序输出字符串

```python
mylist = [1,2,3,4,5,6,7,8,9]
print(mylist[1:3])
print(mylist[:3])
print(mylist[1:])
print(mylist[::2])
print(mylist[::-1])
print(mylist[::1])
```

集合：
1. 集合是元素不可重复的，且无序的，不能通过索引访问，集合中的元素必须是不可变对象，比如数字、字符串、元组等，不能是列表、字典等可变对象
2. 集合用大括号括起来，元素之间用逗号隔开，s = {1,2,3,4,5}
3. 集合可以通过索引访问，索引从0开始，负数索引从-1开始，-1是最后一个元素，-2是倒数第二个元素
4. 集合可以通过切片访问，切片是左闭右开，s[1:3]表示从索引1到索引2，左闭右开
5. 集合可以通过索引修改元素，但是集合是不可变的，所以不能修改集合中的元素
6. 集合可以通过添加元素，删除元素，清空元素，判断元素是否存在等操作

```python
st = {4,3,6,1,2}
print(st)

st = {}
print(type(st))

st = set()
print(type(st))

st = {1, 2, 3, 4, 5}
```

集合的常用操作：
1. len()：返回集合的长度
2. in：判断元素是否在集合中
3. not in：判断元素是否不在集合中
4. add(): 添加元素
5. remove(): 删除元素，如果元素不存在，抛出异常
6. discard(): 删除元素，如果元素不存在，不抛出异常
7. pop(): 随机删除元素，并返回删除的元素
8. clear(): 清空集合
9. union(): 返回两个集合的并集
10. intersection(): 返回两个集合的交集
11. difference(): 返回两个集合的差集

```python
st = {1,2,3,4,5}
print(len(st))
print(1 in st)
print(6 not in st)
st.add(6)
print(st)
st.remove(6)
print(st)
st.discard(6)
print(st)
st.pop() # 随机删除元素，并返回删除的元素
print(st)
st.clear()
print(st)
st1 = {1,2,3,4,5}
st2 = {4,5,6,7,8}
print(st1.union(st2))
print(st1.intersection(st2))
print(st1.difference(st2))
```

字典：
1. 字典是键值对，键必须是不可变对象，值可以是任何对象，字典用大括号括起来，键值对用冒号隔开，键值对之间用逗号隔开，
2. 字典的键是唯一的，如果键重复，后面的值会覆盖前面的值
3. 字典可以通过键访问，值是值，可以通过键访问值，可以通过键删除键值对，可以通过键修改值，可以通过键增加键值对，可以通过键判断键是否存在

```python
dic = {
    "name":"John", # key: value
    "age":18,
    "gender":"male"
}
print(dic)

print(dic["name"])
print(dic.get("name"))
dic["name"] = "Jim"
print(dic)

dic = {
    "name": {
        "first":"John",
        "last":"Doe"
    },
    "grade": {
        "math":90,
        "english":80,
        "science":70
    }
}
print(dic["name"]["first"])
print(dic["grade"]["math"])
print(dic)
```

字典的常用操作：
1. len()：返回字典的长度
2. in：判断键是否存在
3. not in：判断键不存在
4. keys(): 返回字典所有的键
5. values(): 返回字典所有的值
6. items(): 返回字典所有的键值对

```python
dic = {
    "name":"John",
    "age":18,
    "gender":"male"
}
print(len(dic))
print("name" in dic)
print("age" not in dic)
print(dic.keys())
print(dic.values())
print(dic.items())

for key, value in dic.items():
    print(key, value,end="\t")

dic["address"] = "Beijing"
print(dic)
dic.pop("address")
print(dic)
```

五类数据容器的总结：
1. 是否支持下标访问：支持(列表、元组、字符串)、不支持(字典、集合)
2. 是否支持重复元素：支持(列表、元组、字符串)、不支持(集合、字典)
3. 是否支持修改元素：支持(列表、元组、字典)、不支持(字符串、集合)
通用操作：
4. 都支持 for 循环遍历
5. len()：返回容器的长度
6. max()：返回容器的最大值
7. min()：返回容器的最小值
8. sum()：返回容器的和
9. sorted()：返回容器的排序后的副本
10. reversed()：返回容器的逆序迭代器
11. sort()：对容器进行原地排序，返回None。其中可以写 lamda 表达式，如：sorted(mylist, key=lambda x: x[1])。结果都是列表。
12. sorted(reversed=True)：对容器进行原地逆序排序，返回None
不同数据容器的转换：
13. list()：将容器转换为列表，字典转换为列表会丢失键值对，保留键
14. tuple()：将容器转换为元组，字典转换为元组会丢失键值对，保留键
15. str()：将容器转换为字符串
16. set()：将容器转换为集合，字典转换为集合会丢失键值对，保留键，会无序且去重


```python
mylist = [1,2,3,4,5]
print(len(mylist))
print(max(mylist))
print(min(mylist))

import random
random.shuffle(mylist)
print(mylist)
mylist.sort()
print(mylist)
```

函数进阶：
1. 多返回值：函数可以返回多个值，多个值用元组或者列表包裹，如：def func(): return 1,2,3
2. 函数的默认参数：函数的参数可以有默认值，如：def func(a=1): return a
3. 函数的变长参数：函数的参数可以有变长参数，如：def func(*args): return args
4. 参数的四种形式：位置参数、默认参数、变长参数、关键字参数。
5. 不定长参数（位置传递）：在函数定义中，可以使用 * 来定义一个不定长的参数，如：def func(*args): return args。传入的是一个元组，元组中的元素是函数的参数。
6. 不定长参数（关键字传递）：在函数定义中，可以使用 ** 来定义一个不定长的参数，如：def func(**kwargs): return kwargs。
7. 传入的是一个字典，字典中的元素是函数的参数。

```python
def f(a, b):
    return a + b, a -b

print(f(1,2))
x, y = f(1,2)
print(x, y)

def add(*num):
    return sum(num)

print(add(1,2,3,4,5))

def add2(**kwargs):
    return kwargs
print(add2(a=1, b=2, c=3))
```

匿名函数：

2. lambda匿名函数：只能被临时使用一次。
3. lambda语法：lambda 传入参数 : 函数体(一行代码)

Python的文件操作：
1. 文件的编码：默认是 utf-8
2. 文件的打开：file = open(filename, mode, encoding)
    1. r：只读，文件不存在会报错
    2. w：只写，文件不存在会创建，文件存在会清空
    3. a：追加，文件不存在会创建，文件存在会追加
    4. r+：读写，文件不存在会报错
    5. w+：读写，文件不存在会创建，文件存在会清空
    6. a+：读写，文件不存在会创建，文件存在会追加
3. 文件的读取：
    1. file.read() 读取所有内容。也可以指定读取的字节数。多次读取，需要重新打开文件。否则会读取到上次读取的内容。
    2. file.readline() 读取一行内容。可以得到一个列表。可以使用 for 循环遍历。
4. 文件的关闭：file.close()
5. with open() as file: 语句块结束后自动关闭文件。

```python
file = open("test.txt", "r", encoding="utf-8")
print(type(file))
print(file.read())
file.close()

f = open("test.txt", "r", encoding="utf-8")
for line in f: # 读取一行
    # 去除换行符
    line = line.strip()
    print(line,end = "")
```

文件的写入：
1. 打开文件：file = open(filename, mode, encoding)
2. 文件写入：file.write(content) ，写到缓存中。
3. 内容刷新：file.flush() ，只有当刷新之后才会写入到文件中。
4. 关闭文件：file.close()
文件的追加：
5. 打开文件：file = open(filename, "a", encoding)
6. 文件写入：file.write(content) ，写到缓存中。
7. 关闭文件：file.close()

```python
file = open("./out.txt", "w+", encoding="utf-8") # w 模式若文件不存在会创建，存在会清空
file.write("hello world\n")
file.flush()
file.close()

file = open("./out.txt", "a+", encoding="utf-8")
file.write("hello world")
file.flush()    
file.close()
```
