判断语句：(注意缩进)条件是个逻辑表达式，返回True或False
1. if 语句.
2. if else 语句.
3. if elif else ... 语句.
4. 以上判断语句的嵌套使用.


```python
age = 18
if age >= 18:
    print("我已经成年了")
else: # 判断不成立，则执行else语句块
    print("我还未成年")
print("我每天都很开心")

height = float(input("身高多少？(cm)"))
if height > 180:
    print("高")
elif height > 170:
    print("中")
else: # 多个判断都是有序独立的
    print("低")
print("结束")

```

循环语句：(注意缩进)
1. while 循环语句.
2. for in 循环语句.(可以自己设置条件、利用 range() 设置起点和终点和步长、也可以遍历列表、字符串、元组等)
3. 以上语句的嵌套使用.
4. break 语句.退出当前层的循环.
5. continue 语句.退出本次循环，继续下一次循环.
注意变量的作用域。

```python
i = int(1)
while i < 10: # 终止条件
    print(i, end = "\t")
    i += 1 # 步长
```

```python
sum = int(0)
i = int(1)
while i <= 100:
    sum += int(i)
    i += 1
print(sum)
```

```python
i = int(1)
while i <= 9:
    j = int(1)
    while j <= i:
        print(i, "*", j, "=", i*j, end = "\t")
        j += 1
    print()
    i += 1
```

```python
for i in range(1, 10, 1): # range(start, end, step)
    for j in range(1, i+1, 1):
        print(i, "*", j, "=", i*j, end = "\t")
    print()
```

```python
a = str("sdandoiqaaoidnfioqfqino")
cnt = int(0)
for x in a:
    if x == 'a':
        cnt += 1
print(cnt)
```

```python
sum = int(0)
i = int(1)
while True:
    if i % 2 == 0:
        i += 1
        continue
    sum += i
    if sum >= 100:
        print(f"sum = {sum}, step = {i}")
        break
    i += 1
```

函数：
1. 函数定义：def 函数名(参数列表):
2. 函数调用：函数名(参数列表)
3. 函数的嵌套调用：
4. 函数的返回值：return 返回值
5. 函数的参数：
   1. 位置参数：函数名(参数1,参数2,参数3)
   2. 关键字参数：函数名(参数名1=参数值1,参数名2=参数值2)
   3. 默认参数：函数名(参数名1=参数值1,参数名2=参数值2)
   4. 可变参数：函数名()
6. 匿名函数：lambda 参数列表:返回值
7. 函数的递归调用：递归调用自身，注意递归调用的次数。
8. 函数的说明文档："""函数的说明文档"""，主要包括函数的功能、参数、返回值、使用方法等。

```python
def length(s):
    cnt = 0
    for x in s:
        cnt += 1
    return cnt
'''
def 函数名(传入参数):
    函数体
    return 返回值
'''
a = str("asdfasdf")
print(length(a))
```
