Python 中的异常：
1. 语法错误：语法错误，指在 Python 源代码中，语法不正确，无法被解释器识别。
2. 运行时错误：运行时错误，指在 Python 源代码中，语法正确，但是运行时发生错误。
3. 逻辑错误：逻辑错误，指在 Python 源代码中，语法正确，但是运行时没有发生错误，但是结果不正确。
4. 错误处理：错误处理，指在 Python 源代码中，处理运行时错误，让程序继续运行。
5. 异常处理：异常处理，指在 Python 源代码中，处理运行时错误，让程序继续运行。
6. try except else finally：异常处理，指在 Python 源代码中，处理运行时错误，让程序继续运行。

```python
try:
    f = open('abc.txt', 'r') # 尝试打开文件abc.txt
except FileNotFoundError:
    print('文件打开失败')
    
try:
    print(name)
except NameError:
    print('变量未定义')

try:
    print(1/1)
except ZeroDivisionError:
    print('除数不能为0')
else:
    print('没有异常发生')
finally:
    print('无论是否发生异常，都会执行')
```

异常的传递:


```python
def a():
    print("model a start")
    print(1/0)
    
def b():
    print("model b start")
    a()
    
def main():
    try:
        print("main start")
        b()
    except Exception as e:
        print(e)
    
main()
```

自定义模块：
1. 自定义模块，是指在 Python 中定义的函数、类、变量等，这些定义的代码放在一个文件中，这个文件就是一个模块。
2. 在模块中写 if __name__ == "__main__": 的代码，只有当这个模块被直接执行时，才会执行。常被用来测试模块的功能。
3. __all__ 用于 指定模块中哪些函数、类、变量等可以被导入。

```python
import my_modules
print(my_modules.compute(1, 2))
```

Python 的包： package
1. 包是一个目录，其中包含一个或多个 Python 模块。
2. 这个包的目录中必须包含一个名为 __init__.py 的文件。
3. __init__.py 可以是空文件，也可以包含初始化代码。
4. __init__.py 可以用来定义包级别的变量、函数、类等。
5. 可以在 __init__ 中书写 __all__ 变量，用来指定包中哪些模块可以被导入。

```python
import my_package.my_modules1 as k
print(k.print_msg())

from my_package import my_modules2 as b
print(b.print_hello())
```

JSON：
1. JSON 是一种轻量级的数据交换格式。
2. 它基于 JavaScript 对象表示法（JavaScript Object Notation）
3. JSON 的语法与 JavaScript 的对象语法类似。
4. JSON 的值可以是：字符串、数字、布尔值、对象、数组、null
5. Python 中的 json 模块，可以将 JSON 数据转换为 Python 数据。
6. Python 中的 json 模块，可以将 Python 数据转换为 JSON 数据。
7. json 格式例如：
{
  "name": "John",
  "age": 30,
  "married": true,
  "divorced": false,
  "children": ("Ann","Billy"),
  "pets": null,
  "cars": [
    {"model": "BMW 230", "mpg": 27.5},
    {"model": "Ford Edge", "mpg": 24.1}
  ]
}

```python
import json

data = [{"name": "John", "age": 30}, {"name": "Jane", "age": 28}]

json_data = json.dumps(data, ensure_ascii=False) # 将 Python 数据转换为 JSON 数据
print(json_data)

python_data = json.loads(json_data) # 将 JSON 数据转换为 Python 数据
print(python_data)
```

Pyecharts：
1. pip install pyecharts
2. 框架：这是一个可视化库，可以生成各种类型的图表。
   1. 直角坐标系：直角坐标系（Cartesian coordinate system），也称笛卡尔坐标系，是一种二维坐标系，其中 X 轴水平方向，Y 轴垂直方向。直角坐标系是坐标系中常用的坐标表示方式。
   2. 极坐标系：极坐标系（Polar coordinate system），也称球坐标系，是一种二维坐标系，其中 X 轴为角度，Y 轴为半径。极坐标系是坐标系中常用的坐标表示方式。
   3. 地理坐标系：地理坐标系（Geographic coordinate system），也称地理坐标系，是一种三维坐标系，其中 X 轴为经度，Y 轴为纬度，Z 轴为高度。地理坐标系是坐标系中常用的坐标表示方式。

```python
from pyecharts.charts import Line

# 创建一个折线图
line = Line() 
# 添加 x 轴数据
line.add_xaxis(["衬衫", "羊毛衫", "雪纺衫", "裤子", "高跟鞋", "袜子"])
# 添加 y 轴数据
line.add_yaxis("商家A", [5, 20, 36, 10, 75, 90])
# 设置标题和坐标轴名称
line.set_global_opts(title_opts={"title": "折线图示例"}, # 设置标题
                     xaxis_opts={"name": "x 轴"}, # 设置 x 轴名称
                     yaxis_opts={"name": "y 轴"}, # 设置 y 轴名称
                     legend_opts={"is_show": True}, # 显示图例
                     toolbox_opts={"is_show": True}, # 显示工具箱
                     datazoom_opts=[{"type": "slider"}], # 添加数据缩放工具
                     tooltip_opts={"trigger": "axis"}, # 显示提示框
                     )
# 渲染图表
line.render("line.html")
```

基础柱状图：
1. 创建一个柱状图
2. 添加 x 轴数据
3. 添加 y 轴数据
4. 设置标题和坐标轴名称
5. 渲染图表

```python
from pyecharts.charts import Bar

bar = Bar()
bar.add_xaxis(["中国", "美国", "日本", "韩国", "英国", "法国"])
bar.add_yaxis("人口", [1350000000, 312000000, 126000000, 53000000, 51000000, 47000000])
bar.set_global_opts(title_opts={"title": "柱状图示例"}, # 设置标题
                     xaxis_opts={"name": "国家"}, # 设置 x 轴名称
                     yaxis_opts={"name": "人口"}, # 设置 y 轴名称
                     legend_opts={"is_show": True}, # 显示图例
                     toolbox_opts={"is_show": True}, # 显示工具箱
                     datazoom_opts=[{"type": "slider"}], #添加数据缩放工具
)
bar.render("bar.html")
```

基础时间线：
1. 创建一个时间线
2. 添加数据
3. 渲染图表

```python
from pyecharts.charts import Timeline
from pyecharts.options import *
import random

timeline = Timeline()
for i in range(1, 11):
    line = Line() # 创建一个折线图
    line.add_xaxis([f"{i}月{j}日" for j in range(1, 32)]) # 生成日期数据
    line.add_yaxis("销量", [random.randint(1000, 5000) for _ in range(31)]) # 生成随机数据
    line.set_global_opts(title_opts={"title": f"第{i}天"}) # 设置标题
    timeline.add(line, f"{i}月") # 添加时间线
timeline.render("timeline.html")
```
