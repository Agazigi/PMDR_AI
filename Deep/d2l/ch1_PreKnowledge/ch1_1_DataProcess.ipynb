```python
import torch as th
from pprint import pp

from sympy.codegen.ast import float32, int64

# th.arange 函数能够创建一个张量
tensor_a = th.arange(12)
pp(tensor_a)
```

```python
tensor_a.shape # 注意没有括号，shape 是一个属性
```

```python
pp(tensor_a.size) # size 是一个方法，需要有括号
pp(tensor_a.size()) # size 方法
```

```python
tensor_a.numel() # 这个方法能够获取该张量的元素个数
```

```python
tensor_a = tensor_a.reshape((3, 4)) # reshape 函数能够改变一个张量的形状
pp(tensor_a.shape)
pp(tensor_a)
```

```python
tensor_b = th.zeros(20) # 创建一个全 0 的张量
pp(tensor_b)
tensor_b = tensor_b.reshape((2, 2, -1))
pp(tensor_b)
```

```python
tensor_c = th.ones(20) # 全 1
tensor_c
```

```python
tensor_d = th.rand((3, 4)) # 随机数
pp(tensor_d)
```

```python
pp(tensor_d[0])
```

```python
pp(tensor_d[:, 1])
```

```python
tensor_e = th.rand((2, 2))
tensor_f = th.rand((2, 2))
pp(tensor_e)
pp(tensor_f)
```

```python
tensor_e + tensor_f, tensor_e - tensor_f, tensor_e * tensor_f, tensor_e / tensor_f
# 这些都是按照元素进行的运算
```

```python
th.exp(tensor_e)
```

```python
# 重要！！！ cat函数
X = th.zeros((3, 4), dtype=th.float32)
Y = th.ones((3, 4), dtype=th.int64)
Z = th.cat((X, Y)) # 默认是按照第一个轴进行拼接
pp(Z)
```

```python
th.cat((X, Y), dim=0), th.cat((X, Y), dim=1)
```

```python
B = 4
N = 8

feat_1 = th.ones((B, N))
feat_2 = th.zeros((B, N))
th.cat((feat_1, feat_2), dim=0), th.cat((feat_1, feat_2), dim=-1)
```

```python
feat_1 == feat_2
```

```python
feat_1.sum().item() # item 函数能够将张量中的元素拿出来
```
